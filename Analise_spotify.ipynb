```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds_tracks = pd.read_csv('C:/Users/Matheus/Desktop/Datasets/tracks.csv')
ds_tracks.head()
```

```python
#checar valores null

pd.isnull(ds_tracks).sum()
```

```python
ds_tracks.info()
```

```python
#Top 10 musicas menos populares

sorted_ds = ds_tracks.sort_values('popularity', ascending = True).head(10)
sorted_ds
```

```python
ds_tracks.describe().transpose()
```

```python
#Top 10 musicas mais populares

popular_songs = ds_tracks.query('popularity>90', inplace= False).sort_values('popularity', ascending = False)
popular_songs[:10]
```

```python
#Tipagem de dados na coluna "release_date" 

ds_tracks.set_index("release_date", inplace=True)
ds_tracks.index = pd.to_datetime(ds_tracks.index)
ds_tracks.head()
```

```python
#Resultado 
ds_tracks[["artists"]].iloc[18]
```

```python
#Conversao ms para segundos

ds_tracks["duration"] = ds_tracks["duration_ms"].apply(lambda x: round(x/1000))
ds_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
#Valores convertidos para segundos

ds_tracks.duration.head()
```

```python
corr_ds = ds_tracks.drop(["key","mode","explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_ds, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor = "Black")
heatmap.set_title("Mapa de calor de correlação entre variáveis")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
ds_exemplo = ds_tracks.sample(int(0.004*len(ds_tracks)))
```

```python
print(len(ds_exemplo))
```

```python
#Correlação positiva entre "Loudness" e "Energy"

plt.figure(figsize=(10,6))
sns.regplot(data = ds_exemplo, y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy")
```

```python
#Pontos diferentes entre "Popularity" e "Acousticness"

plt.figure(figsize=(10,6))
sns.regplot(data = ds_exemplo, y = "popularity", x = "acousticness", color = "b").set(title = "Popularity vs Acousticness")
```

```python
ds_tracks['dates'] = ds_tracks.index.get_level_values('release_date')
ds_tracks.dates = pd.to_datetime(ds_tracks.dates)
years = ds_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect = 2, height= 5, kind="hist").set(title="Numero de musicas por ano (Limitado até 2020)")
```

```python
#Comparação entre a duração das musicas com o decorrer dos anos

total_dr = ds_tracks.duration
figure_dim = (18,7)
fig, ax = plt.subplots(figsize = figure_dim)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Ano vs Duração das musicas")
plt.xticks(rotation=60)
```

```python
#Visualização com grafico de linha entre Ano vs Duração

total_dr = ds_tracks.duration
sns.set_style(style="whitegrid")
fig_dim = (10,5)
fig, ax = plt.subplots(figsize = fig_dim)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title = "Ano vs Duração")
plt.xticks(rotation=60)
```
